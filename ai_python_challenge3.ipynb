```python
import pandas as pd
df = pd.read_csv("/content/reviews.csv")
```

```python
import os
from google.colab import userdata
os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')
```

```python
from groq import Groq

client = Groq()
```

```python
reviews = df['reviewText']
```

```python
def classificador_de_reviews(reviews):
  respostas = []
  for review in reviews:
    resposta = ""
    completion = client.chat.completions.create(
        model="openai/gpt-oss-120b",
        messages=[
          {
            "role": "user",
            "content": f'Classifique o seguinte texto de revisao como Positivo, Negativo ou Neutro: {review}'
          }
        ],
        temperature=1,
        max_completion_tokens=8192,
        top_p=1,
        reasoning_effort="medium",
        stream=True,
        stop=None
    )
    for chunk in completion:
      content = chunk.choices[0].delta.content
      if content:
          resposta += content
    respostas.append(resposta)
  return respostas
```

```python
reviews_classificadas = classificador_de_reviews(reviews)
```

```python
df["sentimento"] = reviews_classificadas
```
